```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
```

```python
data = pd.read_csv('./train.csv')
# pd.isna(data.SalePrice).sum() #计算SalePrice一栏的nan数量
```

```python
data.dropna(axis=0, subset=['SalePrice'], inplace=True)
y = data.SalePrice
X = data.drop(['SalePrice'], axis=1).select_dtypes(exclude=['object'])
```

```python
data
```

```python
y
```

```python
train_X, test_X, train_y, test_y = train_test_split(X.values, y.values, test_size=0.25)
my_imputer = SimpleImputer()
train_X = my_imputer.fit_transform(train_X)
test_X = my_imputer.transform(test_X)
```

```python
import xgboost
from xgboost import XGBRegressor

my_model = XGBRegressor()
# Add silent=True to avoid printing out updates with each cycle
my_model.fit(train_X, train_y, verbose=False)
```

```python
# make predictions
predictions = my_model.predict(test_X)

from sklearn.metrics import mean_absolute_error
print("Mean Absolute Error : " + str(mean_absolute_error(predictions, test_y)))
```

```python
# early_stopping_rounds参数在构造器中传入
my_model = XGBRegressor(n_estimators=1000, early_stopping_rounds=5)
my_model.fit(train_X, train_y, eval_set=[(test_X, test_y)], verbose=False)
```

```python
my_model = XGBRegressor(n_estimators=1000, learning_rate=0.05)
my_model.fit(train_X, train_y, early_stopping_rounds=5, 
             eval_set=[(test_X, test_y)], verbose=False)
```
